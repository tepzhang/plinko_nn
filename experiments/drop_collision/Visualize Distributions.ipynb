```python
%matplotlib inline
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
import numpy as np
import pandas as pd
from plotnine import *
```

```python
simulations = pd.read_feather('simulations.feather')
collisions = simulations[['obs_n_sides', 'obs_x', 'obs_y', 'obs_rot', 'vx_00', 'vy_00', 'vy_01', 'vx_01']]
collisions['angle_01'] = np.ar
```

```python
(ggplot(collisions)
 + aes(x = 'vy_01')
 + geom_histogram()
)
```

```python
df = collisions[collisions.obs_n_sides == 3]
plt.hist2d(df.vx_01, df.vy_01, bins=40)
```
